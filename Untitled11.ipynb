```python
a =np.array([1,6,3,4,5,6,7])
b=np.array([1,2,8,4,5,6,7])
```

```python
g =np.array([3,56,4,34,63,78,23])
h=np.array([22,3,6,64,66,23,67])
plt.xlabel('g')
plt.ylabel('h')
plt.plot(g,h)
plt.show()
```

```python
x= [5,8,10]
y =[12,16,6]
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```
