1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
`You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#lets frist connect the web driver
driver = webdriver.Chrome("chromedriver.exe")
```

```python
#accessing the webdriver
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(2)


#creating empty list
most_viewed_video = { }
most_viewed_video['Rank']=[]
most_viewed_video['Video Name']=[]
most_viewed_video['Uploader']=[]
most_viewed_video['Views in billion']=[]
most_viewed_video['Upload date']=[]
most_viewed_video['Note']=[]


#video rank
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]"):
    most_viewed_video['Rank'].append(i.text)
    
#video Name    
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]"):
    most_viewed_video['Video Name'].append(i.text)
    
#Uploader    
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]"):
    most_viewed_video['Uploader'].append(i.text)
    
#Views in billions    
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]"):
    most_viewed_video['Views in billion'].append(i.text)
    
#upload Date    
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]"):
    most_viewed_video['Upload date'].append(i.text)

#Note    
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[6]"):
    most_viewed_video['Note'].append(i.text)
    
    
Most_viewed_youtube_video = pd.DataFrame(most_viewed_video)
Most_viewed_youtube_video[:30]
    
```

2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code

```python
#lets frist connect the web driver
driver = webdriver.Chrome("chromedriver.exe")

#accessing the web page 
driver.get("https://www.bcci.tv")
time.sleep(3)


#clicking on internation drop down
International = driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']") 
International.click()
time.sleep(3)

#clicking on Fixtures
Fixtures = driver.find_element_by_xpath("//div[@class='drop-down__options']/ul/li/a")
Fixtures.click()
time.sleep(3)


Criket_Matches = { }
Criket_Matches['Match Title']=[]
Criket_Matches['Series']=[]
Criket_Matches['Place']=[]
Criket_Matches['Date']=[]
Criket_Matches['Time']=[]
Criket_Matches['Month']=[]



#scraping match title 
for i in driver.find_elements_by_xpath("//Strong[@class='fixture__name fixture__name--with-margin']"):
    Criket_Matches['Match Title'].append(i.text)
    
#scarping the series 
for i in driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']"):
    Criket_Matches['Series'].append(i.text)
 
#scarping the Date
for i in driver.find_elements_by_xpath("//span[@class='fixture__date']"):
    Criket_Matches['Date'].append(i.text)
    
#scarping the Month
for i in driver.find_elements_by_xpath("//span[@class='fixture__month']"):
    Criket_Matches['Month'].append(i.text)
    
#scarping the Time   
for i in driver.find_elements_by_xpath("//span[@class='fixture__time']"):
    Criket_Matches['Time'].append(i.text)    
    
#scarping the place
for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):
    Criket_Matches['Place'].append(i.text)
    
#creating the data frame 
Criket_Matches = pd.DataFrame(Criket_Matches)
Criket_Matches.head(10)

```

```python


url ="https://github.com/"
driver.get(url)
time.sleep(7)


explore_opt = driver.find_element_by_xpath("//ul[@class='d-lg-flex list-style-none']/li[4]")
action = ActionChains(driver)
action.click_and_hold(explore_opt).perform()


trend_opt = driver.find_element_by_xpath("//ul[@class='list-style-none mb-3'][2]/li[3]/a")
action.click(trend_opt).perform()


trending_repo = {}
trending_repo['Repository_Title'] = []
trending_repo['Repository_Description'] = []
trending_repo['Contributors_Count'] = []
trending_repo['Language_Used'] = []


titles = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in titles:
    if i.text == " ":
        trending_repo['Repository_Title'].append("-")
    else:
        trending_repo['Repository_Title'].append(i.text)
time.sleep(2)

#description
count = 1
for i in range(len(driver.find_elements_by_xpath("//article[@class='Box-row']"))):
    
    try:
        descriptions = driver.find_element_by_xpath(f"//article[@class='Box-row'][{count}]/p[1]")
        trending_repo['Repository_Description'].append(descriptions.text) 
        
    except NoSuchElementException:
        trending_repo['Repository_Description'].append("-")
    
    count += 1

#contributor_counts
contributor_counts = driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']/a[2]")
for i in contributor_counts:
    if i.text == "":
        trending_repo['Contributors_Count'].append("-")
    else:
        trending_repo['Contributors_Count'].append(i.text)
time.sleep(2)

#language
lan_count = 1
for i in range(len(driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']"))):
    try:
        langu = driver.find_element_by_xpath(f"//article[@class='Box-row'][{lan_count}]/div[2]/span/span[2]")
        trending_repo['Language_Used'].append(langu.text) 
        
    except NoSuchElementException:
        trending_repo['Language_Used'].append("-")
    lan_count += 1
time.sleep(2)


#creating dataframe
Trending_repositories = pd.DataFrame(trending_repo)
Trending_repositories.to_csv("Trending repositories.csv")
Trending_repositories.head()
```

```python
# Accessing chrome browser
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(3)

driver.get("https://archive.ics.uci.edu/")
time.sleep(4)


action = ActionChains(driver)
view = driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td/span/b/a")
action.click(view).perform()
time.sleep(5)

ML_repo = {}
ML_repo['Dataset name']= []
ML_repo['Data type'] = []
ML_repo['Task'] = []
ML_repo['Attribute type'] = []
ML_repo['No of instances'] = []
ML_repo['No of attribute'] = []
ML_repo['Year'] = []


# Scraping the required detail

name = driver.find_elements_by_xpath("//p[@class='normal']/b/a")
for i in name:
    if len(ML_repo['Dataset name']) < 586:
        ML_repo['Dataset name'].append(i.text)
    else:
        pass
     
time.sleep(1)
for i in range(2,588):
    dtype = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[2]/p")
    if dtype.text == " ":
        ML_repo['Data type'].append("-")
    else:
        ML_repo['Data type'].append(dtype.text)
    
    task = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[3]/p")
    if task.text == " ":
        ML_repo['Task'].append("-")
    else:
        ML_repo['Task'].append(task.text) 
        
    type = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[4]/p")
    if type.text == " ":
        ML_repo['Attribute type'].append("-")
    else:
        ML_repo['Attribute type'].append(type.text)
        
    instance = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[5]/p")
    if instance.text == " ":
        ML_repo['No of instances'].append("-")
    else:
        ML_repo['No of instances'].append(instance.text) 
        
    num_attribute = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[6]/p")
    if num_attribute.text == " ":
        ML_repo['No of attribute'].append("-")
    else:
        ML_repo['No of attribute'].append(num_attribute.text)        
        
    year = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[7]/p")
    if year.text == " ":
        ML_repo['Year'].append("-")
    else:
        ML_repo['Year'].append(year.text)
time.sleep(1)


#Creating Dataframe
ML_repo = pd.DataFrame(ML_repo)
ML_repo.to_csv("ML_repo.csv")
ML_repo.head(20)
```
